```python
total_revenue = 0

for sale in sales_data:
    total_revenue += sale["quantity"] * sale["price_per_item"]

print("Total Revenue (INR):", total_revenue)
```

```python
high_value_orders = []

for sale in sales_data:
    order_value = sale["quantity"] * sale["price_per_item"]
    if order_value > 2000:
        high_value_orders.append(sale)

print("High Value Orders (Order Value > ₹2000):")
for order in high_value_orders:
    print(order)
```

```python
item_quantity = {}

for sale in sales_data:
    item = sale["item"]
    quantity = sale["quantity"]

    if item in item_quantity:
        item_quantity[item] += quantity
    else:
        item_quantity[item] = quantity

highest_selling_item = max(item_quantity, key=item_quantity.get)

print("Item with Highest Total Quantity Sold:", highest_selling_item)
print("Total Quantity Sold:", item_quantity[highest_selling_item])
```

```python
def apply_discount(orders, discount_rate):
    discounted_orders = []

    for order in orders:
        # Create a copy of the order to avoid modifying the original
        discounted_order = order.copy()

        # Apply discount to price_per_item
        discounted_order["price_per_item"] = round(
            discounted_order["price_per_item"] * (1 - discount_rate), 2
        )

        discounted_orders.append(discounted_order)

    return discounted_orders
```

```python
discounted_sales = apply_discount(sales_data, 0.10)

print("Original Order Price:", sales_data[0]["price_per_item"])
print("Discounted Order Price:", discounted_sales[0]["price_per_item"])
```

```python
discount_rate = 0.10  # Global variable

def calculate_discount(price):
    discount_amount = price * discount_rate  # Local variable
    return discount_amount

print("Discount:", calculate_discount(1000))
```

```python
counter = 0  # Global variable

def increment_counter():
    global counter
    counter += 1

increment_counter()
print("Counter value:", counter)
```

```python
class OrderIterator:
    def __init__(self, orders):
        self.orders = orders
        self.index = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.index < len(self.orders):
            order = self.orders[self.index]
            self.index += 1
            return order
        else:
            raise StopIteration
```

```python
order_iterator = OrderIterator(sales_data)

for order in order_iterator:
    print(order)
```
